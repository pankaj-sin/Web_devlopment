```python
#importing the python libraries - pandas and numpy 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing & reading the dataset
#This creates a dataframe from the CSV file:
df=pd.read_csv("assignment .csv",index_col='Date',parse_dates=True)
```

```python
df
```

```python
df.columns
```

```python
#Checking if any column has any empty value:
df.isna().sum()
```

```python
#dropping the empty data entries:
df.dropna(inplace=True)
df.isna().sum()
```

```python
df
```

```python
df['SMoving_Average7'] = df['Close'].rolling(window=7).mean()
```

```python
df.head(2000)
```

```python
df['SMoving_Average14'] = df['Close'].rolling(window=14).mean()
```

```python
df.head(3000)
```

```python
#To mprove our plot appearence ,we can use a pre-defined style 
plt.style.use('fivethirtyeight')
```

```python
#Our Chart size
plt.figure(figsize = (13,7))

#Plotting ' Close' Price and SMA_7 lines
plt.plot(df.Close,label=' Close',linewidth=3)
plt.plot(df.SMoving_Average,label='7 day Rolling SMA',linewidth=1.5)

#Adding title, labels on the axis:
#Legend on the left-upper corner
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.title('Pot of  Close v/s Simple_MA_7')
plt.legend()
```

```python
df.drop(['SMoving_Average'],axis=1,inplace=True)
```

```python
df.head(89)
```

```python
df.to_csv("assignment .csv")
```

```python
#this selects the ' Close' column
close=df['Close']

#This converts the date strings in the index into pandas datetime format:
close.index=pd.to_datetime(close.index)
close
```

```python
#Plot the price series for visual inspection:
close.plot()
```

```python
sma_7=close.rolling(window=7).mean()
sma_7
```
